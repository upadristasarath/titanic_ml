```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataFrame = pd.read_csv("train.csv")
```

```python
# crosstab - compares two variables
# How many males and females survived
pd.crosstab(dataFrame['Sex'], dataFrame['Survived'])
# O/p - 0 means didn't survice, 1 means survived
```

```python
 dataFrame.columns
```

```python
pd.crosstab(dataFrame['Pclass'], dataFrame['Survived'])
```

```python
# Visualization using seaborn and matplotlib 
ax = sns.countplot(x="Sex", hue="Survived", palette="Set1", data=dataFrame)
ax.set(title="Survivors according to sex", xlabel="Sex", ylabel="Total")
plt.show()
```

```python
# sns.factorplot(x="Pclass", y="Survived", hue="Sex", data=dataFrame, aspect=0.9, size=3.5)

sns.catplot(x="Pclass", y="Survived", hue="Sex", data=dataFrame, aspect=0.9, size=3.5)
```

```python
sns.catplot(x="Embarked", y="Survived", hue="Sex", data=dataFrame, aspect=0.9, size = 3.5)
```
