```python
# Machine learning alogs require everything in numbers so we should change to numbers
import pandas as pd

df = pd.read_csv("train.csv")
```

```python
df.sample(20)
```

```python
def get_title(name):
    if "." in name:
        return name.split(",")[1].split(".")[0].strip()
    else:
        return "No Titles for the Name"
```

```python
titles = set([x for x in df.Name.map(lambda x: get_title(x))])
```

```python
print(titles)
```

```python
def short_titles(x):
    title = x["Title"]
    if title in ['Capt', 'Col', 'Major']:
        return 'Officer'
    elif title in ['Jonkheer', 'Don',  "the Countess", "Done", "Lady", "Sir"]:
        return "Royality"
    elif title == "Mme":
        return "Mrs"
    elif title in ["Mlle", "Ms"]:
        return "Miss"
    else:
        return title
```

```python
df["Title"] = df["Name"].map(lambda x: get_title(x))
```

```python
df.sample(20)
```

```python
df['Title'] = df.apply(short_titles, axis=1)
```

```python
df['Title'].value_counts()
```

```python
# del df['Name']
df.drop("Name", axis=1, inplace=True)
```

```python
df.sample(10)
```

```python
df.Sex.replace(('male', 'female'), (0,1), inplace=True)
```
