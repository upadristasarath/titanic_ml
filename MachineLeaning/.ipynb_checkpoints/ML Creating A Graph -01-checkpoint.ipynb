```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph to show the effects of X on Y")
plt.show()
```

```python
ad_budget = [1,10,50,100,1000] # x - axis
sales = [10,100,500,1000,10000] # y-axis

plt.plot(ad_budget, sales)
plt.xlabel("Amount spent on fb ads ($)")
plt.ylabel("Amount of Sales ($)")
plt.title("Graph to show the effects of ad budgetting on sales")
plt.show()
```
