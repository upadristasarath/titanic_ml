```python
import matplotlib.pyplot as plt
# Effect of time spent walking (hours) on the distance travelled (miles)
time_spent_walking = [1,2,3,4,5]
distance = [2,4,6,8,10]

plt.plot(time_spent_walking, distance)
plt.xlabel("Time spent on walking (hours)")
plt.ylabel("Distance travelled (miles)")
plt.title("Graph to show the effects of time spent on walking on distance travelled")
plt.show()
```

```python
# Effect of car age (years) on price ($)
car_age=[1,2,5,10,30]
price =[30000, 25000, 18000,10000,4000]

plt.plot(car_age, price)
plt.xlabel("Age of the car (years)")
plt.ylabel("Price of the car ($)")
plt.title("Graph to show the effects of age of the car on the price")
plt.show()
```
