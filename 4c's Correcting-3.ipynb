```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataFrame = pd.read_csv("train.csv")
```

```python
graph = sns.FacetGrid(dataFrame, col="Survived")
graph.map(plt.hist, "Fare", bins=20)
```

```python
# dataFrame.loc[dataFrame['Fare'] > 400, "Fare"] = 1000 

# made corrections using median
dataFrame.loc[dataFrame['Fare'] > 400, "Fare"] = dataFrame["Fare"].median()
#dataFrame["Fare"].median()
```

```python
graph_age = sns.FacetGrid(dataFrame, col="Survived")
graph_age.map(plt.hist, "Age", bins=20)
```

```python
# survived age is 80(very few). Want to correct this my changing to 62/ median
# dataFrame.loc[dataFrame["Age"] > 70, "Age"] = dataFrame["Age"].median()
dataFrame.loc[dataFrame["Age"] > 70, "Age"] = 62
```
